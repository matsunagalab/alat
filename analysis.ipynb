```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
traj = md.load('run.dcd', top='init.psf')
traj = traj.atom_slice(traj.topology.select("protein"))
```

```python
phi = md.compute_phi(traj)[1][:, 0]
psi = md.compute_psi(traj)[1][:, 1]
```

```python
phi
```

```python
plt.scatter(phi * 180.0 / np.pi, psi * 180.0 / np.pi)
plt.xlabel("phi [degree]")
plt.ylabel("psi [degree]")
plt.savefig("analysis.png", dpi=350)
```
